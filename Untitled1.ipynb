```python
import numpy as np
import matplotlib.pyplot as plt 
im=plt.imread('a.png')
type(im)
im.shape
```

```python
im[100,100,:] #100,100 konumundaki renk deÄŸerleri
```

```python
plt.imshow(im)
plt.show() #resmi yazdÄ±rÄ±yor
```

```python
m,n,p = im.shape #resmin konum ve boyut deÄŸerlerini sÄ±rasÄ±yla m,n,p'ye atar.
m,n,p
```

```python

#Resmi siyah beyaz yapmak iÃ§in: her pixel'i RGB'nin ortalamasÄ±na eÅŸitliyoruz.

new_im=np.zeros((m,n),dtype=float) 

for i in range (m):
    for j in range (n):
        s=(im[i,j,0]+im[i,j,1]+im[i,j,2])/3
        new_im[i,j]=s

plt.imshow(new_im,cmap='gray')
plt.show() 
plt.imsave('siyahbeyaz.png',new_im,cmap='gray')
```

```python
#Resmi 90 derece sola dÃ¶ndÃ¼rmek iÃ§in:

new_im=np.zeros((n,m),dtype=float) 

for i in range (m):
    for j in range (n):
        s=(im[i,j,0]+im[i,j,1]+im[i,j,2])/3
        new_im[j,i]=s

plt.imshow(new_im,cmap='gray')
plt.show() 
plt.imsave('doksanderece.png',new_im,cmap='gray')
```
